```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import math
from sympy import *


eps = 0.0001
x1 = 42
x2 = 11
h = 0


def f(x1, x2): 
    return 100 * (x2-x1**3)**2 + (1-x1)**2

def grad_x1(x1, x2): #аналитический метод наождения частной производной
    return -600 * (x1 ** 2) *(x2 - x1 ** 3) + 2  * x1 - 2
   
def grad_x2(x1,x2): #аналитический метод наождения частной производной
    return -200 * x1 ** 3 + 200 * x2
#print (grad_x2(x1,x2))


def ret_grad(x1, x2, grad_x1, grad_x2): #находим шаг   
    k = Symbol("k") 
    x1_2 = x1 - k * grad_x1(x1, x2)
    x2_2 = x2 - k * grad_x2(x1, x2)
    pr = diff ((x1_2 ** 2 + x2_2 ** 2 + (3 * x1_2 * x2_2) / 2 ))
    k = solve(pr)
    b = [float(i) for i in k]

    k1 = b[0]
    x1_2 = x1 - k1 * grad_x1(x1, x2) 
    x2_2 = x2 - k1 * grad_x2(x1, x2)

#    print ('Шаг: ', k1) 
    return k1
#print ('Шаг: ',ret_grad(x1, x2, grad_x1, grad_x2))

def chast_diff_x1x2 (x1, x2, x1_0, x2_0): #численный метод нахождения частной производной
    x1, x2 = symbols('x1 x2')
    chast_diff_x1 = diff(100 * (x2-x1**3)**2 + (1-x1)**2, x1)
    chast_diff_x2 = diff(100 * (x2-x1**3)**2 + (1-x1)**2, x2)    
#    print ('Частная призводная по х1: ', chast_diff_x1)
#    print ('Частная призводная по х2: ', chast_diff_x2)
    
    chast_diff_x1 = chast_diff_x1.subs ({x1: x1_0, x2: x2_0}) #значение градиента
    chast_diff_x2 = chast_diff_x2.subs ({x1: x1_0, x2: x2_0})
    
    return chast_diff_x1,chast_diff_x2 

#print ('Значение градиента: ',chast_diff_x1x2 (x1, x2, x1, x2))
 
def new_x1x2 (x1, x2, ret_grad, chast_diff_x1x2):
    x1_2 = x1 - ret_grad(x1, x2, grad_x1, grad_x2) * chast_diff_x1x2(x1, x2, x1, x2)[0]
    x2_2 = x2 - ret_grad(x1, x2, grad_x1, grad_x2) * chast_diff_x1x2(x1, x2, x1, x2)[1]
    #print (f(x1_2, x2_2))
    return (x1_2, x2_2)
    
#print ('Значение новый переменных: ', new_x1x2( x1, x2, ret_grad, chast_diff_x1x2))

#a = new_x1x2(x1, x2, ret_grad, chast_diff_x1x2)
#print (a)
#print(f(a[0],a[1]))

def task (x1, x2, ret_grad, chast_diff_x1x2, new_x1x2,f):
    k=1
    c=0
    while (True):
        print('k =', k)
        print(' Начальная точка =', '(', x1 , ',', x2, ')')
        print(' Значение функции в начальной точке = ' , f(x1,x2))
        chast_diff_x1x2(x1, x2, x1, x2)
        c +=2
        print(' Градиент:' , chast_diff_x1x2(x1, x2, x1, x2))
        ret_grad(x1, x2, grad_x1, grad_x2)
        print(' Шаг = ', ret_grad(x1, x2, grad_x1, grad_x2))
        a = new_x1x2(x1, x2, ret_grad, chast_diff_x1x2)
        c +=4
        print(' Новая точка:' , new_x1x2(x1, x2, ret_grad, chast_diff_x1x2))
        print(' Значение функции в новой точке:', f(a[0], a[1]))
        if abs ( f(a[0], a[1]) - f(x1, x2)) < eps :
            break
        
        x1=a[0]
        x2=a[1]
        k+=1
        c+=2
    print('Ответ: ')
    return x1, x2, f(x1,x2), c
        
print(task (x1, x2, ret_grad, chast_diff_x1x2, new_x1x2,f))
```
