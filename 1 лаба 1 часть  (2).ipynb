Зададим функцию

```python
def f(x):
    return (x-15)**2+5
```

```python
a = 2
b = 200
e=0.01
d=0.001
```

Метод дихотомии (деления отрезка пополам)

```python
def dichotomy(f, a, b):
    k = 1
    count = 0
    while (b-a)>e:
        middle = (a+b)/2
        x1 = middle-d
        x2 = middle+d 
        print('k =', k) 
        print(' Интервал нахождения минимума функции:', '[', a, ',', b, ']', '\n Длина интервала: ', b-a)
        print(' Середина интервала =', middle, '\n Значение функции в середине интервала =', f(middle))
        print(' x1 =', x1, '\n Значение функции в x1 =', f(x1))        
        print(' x2 =', x2, '\n Значение функции в x2 =', f(x2), '\n')        
        if f(x1)>f(x2):
            a = x1
        else:
            b = x2 
        k += 1
        count +=2
    print ('Ответ:\n xmin =', middle)
    print (' значение функции в xmin =', f(middle), '\n') 
    return a, b, b-a, count

print(dichotomy(f, a, b))    
```

```python
import math 

def golden(a, b, f):
    k = 1
    c = 0
    x1 = a + 0.381966011*(b-a)
    x2 = a + 0.618003399*(b-a)
    while (b - a) > e:
        print('k =', k) 
        print(' Интервал нахождения минимума функции:', '[', a, ',', b, ']', '\n Длина интервала: ', b-a)
        print(' x1 =', x1, '\n Значение функции в x1 =', f(x1))
        print(' x2 =', x2, '\n Значение функции в x2 =', f(x2), '\n')
        if f(x1) < f(x2):
            b = x2
            x2=x1
            x1=a + 0.381966011*(b-a)
        else:
            a = x1  
            x1=x2
            x2 = a + 0.618003399*(b-a)
        k += 1
        c += 2
        
    print ('Ответ:\n xmin =', (b+a)/2)
    print (' значение функции в xmin =', f((b+a)/2), '\n') 
    return a, b, b-a, c 

print(golden(a, b, f))
```

Метод Фибоначчи

```python
from math import sqrt 
def F(n): 
    return ((1+sqrt(5))**n-(1-sqrt(5))**n)/(2**n*sqrt(5))
```

```python
Fmax=((b-a)/e)
F1 = 1
F2 = 1
Fn = 0
n = 2
while (Fmax > Fn):
    Fn = F1 + F2
    F1 = F2
    F2 = Fn
    n += 1
print(n)
```

```python
def fibonachchi(a, b, f):
    k=1
    count =0
    n=21
    x1 = a + (F(n)/F(n+2))*(b-a)
    x2 = a + (F(n+1)/F(n+2))*(b-a)
    while k!=n+1:
        print('k =', k) 
        print(' Интервал нахождения минимума функции:', '[', a, ',', b, ']', '\n Длина интервала: ', b-a)
        print(' x1 =', x1, '\n Значение функции в x1 =', f(x1))
        print(' x2 =', x2, '\n Значение функции в x2 =', f(x2), '\n')
        if f(x1)>f(x2):
            a=x1
            x1=x2
            x2=a+(F(n-k+2)/F(n+2))*(200-2)
        else:
            b=x2
            x2=x1
            x1=a+(F(n-k+1)/F(n+2))*(200-2)
        k += 1 
        count +=2
    
    print ('Ответ:\n xmin =', (b+a)/2)
    print (' значение функции в xmin =', f((b+a)/2), '\n') 
    print ('x1 =', x1, 'x2 =', x2)
    return a, b, b-a , count
    
print(fibonachchi(a, b, f))     
```
