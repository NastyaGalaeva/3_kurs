Зададим функцию

```python
def f(x):
    return (x-15)**2+5
```

```python
a = 2
b = 200
```

Метод дихотомии (деления отрезка пополам)

```python
def dichotomy(f, a, b, e):
    k = 1
    while (b-a)>0.01:
        middle = (a+b)/2
        x1 = middle-e/2 
        x2 = middle+e/2 
        print('k =', k) 
        print(' Интервал нахождения минимума функции:', '[', a, ',', b, ']', '\n Длина интервала: ', b-a)
        print(' Середина интервала=', middle, '\n Значение функции в середине интервала=', f(middle))
        print(' x1=', x1, '\n Значение функции в x1=', f(x1))
        print(' x2=', x2, '\n Значение функции в x2=', f(x2), '\n')
        if f(x1)>f(x2):
            a = x1
        else:
            b = x2 
        k += 1
    return a, b, b-a

print(dichotomy(f, a, b, e))    
```
